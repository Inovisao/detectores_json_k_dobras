### 1. Script para checar estrutura e endere√ßamento de diret√≥rios. üìÑ

```python
import os

def listar_estrutura_pasta(diretorio, nivel=0, caminho_atual=""):
    for item in os.listdir(diretorio):
        caminho_item = os.path.join(diretorio, item)
        caminho_atual_item = os.path.join(caminho_atual, item)

        if os.path.isdir(caminho_item):
            print("|   " * nivel + f"|--- {item}: {caminho_atual_item}")

            listar_estrutura_pasta(caminho_item, nivel + 1, caminho_atual=caminho_atual_item)

if __name__ == "__main__":
    diretorio_base = '/content'
    print(f"Organiza√ß√£o da estrutura de pastas em {diretorio_base}:")
    listar_estrutura_pasta(diretorio_base)
```

### 2. Script para mover arquivos .txt e .jpg de um diret√≥rio para outro.

```python
import os
import shutil

def mover_arquivos(source_directory, destination_directory):
    # Verifica se os diret√≥rios de origem e destino existem
    if not os.path.exists(source_directory):
        print(f"Diret√≥rio de origem '{source_directory}' n√£o encontrado.")
        return

    if not os.path.exists(destination_directory):
        os.makedirs(destination_directory)
        print(f"Diret√≥rio de destino '{destination_directory}' criado.")

    # Itera sobre os arquivos no diret√≥rio de origem
    for filename in os.listdir(source_directory):
        if filename.endswith(".jpg") or filename.endswith(".txt"):
            origem = os.path.join(source_directory, filename)
            destino = os.path.join(destination_directory, filename)

            # Move o arquivo
            shutil.move(origem, destino)
            print(f"Arquivo {filename} movido para {destination_directory}")

if __name__ == "__main__":
    # Substitua esses caminhos pelos seus diret√≥rios de origem e destino
    diretorio_origem = "/content/yolo/"
    diretorio_destino = "/content/yolo/yolo_dataset/"

    mover_arquivos(diretorio_origem, diretorio_destino)
```

### **4. Classes que ir√£o definir fun√ß√µes para que o arquico coco.json obtido a partir do Download do dataset do Roboflow sejam processadas pelo SAM.**



Alguns pontos importantes:

A classe **MyRoboFlow** √© respons√°vel por interagir com a API do Roboflow para baixar datasets de projetos espec√≠ficos. No m√©todo get, ela utiliza a biblioteca Roboflow para obter o √∫ltimo conjunto de dados de um projeto espec√≠fico, e o salva no formato COCO.

A classe **ExtractCoco** √© projetada para extrair informa√ß√µes relevantes de um arquivo COCO, como caixas delimitadoras, r√≥tulos e √°reas. Essas informa√ß√µes s√£o organizadas em um formato mais conveniente para posterior processamento. O m√©todo extract_dic converte as informa√ß√µes do COCO em um formato que pode ser facilmente utilizado em outras partes do c√≥digo.

A classe **ExecuteSam** utiliza o modelo SAM para prever m√°scaras a partir das caixas delimitadoras extra√≠das anteriormente. O m√©todo convertCOCO2Sam faz a convers√£o do formato COCO para o formato de entrada esperado pelo modelo SAM. As m√°scaras resultantes s√£o salvas em diret√≥rios espec√≠ficos, e a sobreposi√ß√£o dessas m√°scaras na imagem original √© salva em outro diret√≥rio.

A classe **Files** √© uma utilidade para listar todos os arquivos JSON presentes em um diret√≥rio. O m√©todo extractJson chama o m√©todo lists para obter uma lista de todos os arquivos JSON presentes no diret√≥rio especificado.

Certifique-se de alterar as informa√ß√µes relevantes, como as chaves da API do Roboflow, o caminho do modelo SAM, e os diret√≥rios de salvamento de m√°scaras e imagens combinadas.

```python
class MyRoboFlow(object):
    def __init__(self, api_key='U7msdHq9U4FEO3233smk', user='mariah.ojeda', project='inseto-dixzb'):
        print('Configure ..')
        self.api_key = api_key
        self.user = user
        self.project = project

    def get(self, to=None):
        try:
            rf = Roboflow(api_key=self.api_key)

            # List projects in the workspace
            projects = rf.workspace(self.user).projects()
            project = next((p for p in projects if p.name == self.project), None)

            if project:
                # Download the latest version of the project
                dataset = project.latest_version().download("coco-mmdetection")
                local = str(dataset.location)

                if to is not None:
                    shutil.move(local, to)

                return local.split(os.sep)[-1]
            else:
                raise ValueError(f"Project '{self.project}' not found in the specified workspace.")
        except Exception as error:
            print(f"Error: {error}")
```

```python
class ExtractCoco(object):

    def __init__(self, path):
        try:
            assert os.path.exists(path), 'File [' + path + '] not exists!'
            self.coco = COCO(path)
            self.dic_labels = {}
            for element in self.coco.cats.values():
                self.dic_labels[int(element['id'])] = element['name']
        except ValueError as error:
            print(error)

    def extract_dic(self, prefix='.'):
        try:
            imgs = []
            for key in self.coco.imgs.keys():
                annotations = self.coco.loadAnns(self.coco.getAnnIds(imgIds=key))
                self.size = [self.coco.imgs[key]['width'], self.coco.imgs[key]['height']]
                boxes, labels, areas, iscrowd, masks, image_id = self.getAnnotations(annotations)
                target = {}
                target['boxes'] = boxes
                target['labels'] = labels
                target['image_id'] = image_id
                target['area'] = areas
                target['iscrowd'] = iscrowd
                imgs.append([os.path.join(prefix, self.coco.imgs[key]['file_name']), target, self.size])
            return imgs
        except ValueError as error:
            print(error)

    def getAnnotations(self, annotations, normalize=False):
        try:
            boxes = []
            labels = []
            areas = []

            for annotation in annotations:
                # Getting all boxes
                box = [float(val) for val in annotation.get('bbox')]
                box[2] += box[0]
                box[3] += box[1]
                box[0] = int((box[0] + box[2]) / 2)
                box[1] = int((box[1] + box[3]) / 2)
                boxes.append(box[:2])
                # Getting all labels
                labels.append(annotation.get('category_id'))
                # Getting all areas
                areas.append(annotation.get('area'))
            boxes = torch.FloatTensor(boxes)
            labels = torch.as_tensor(labels, dtype=torch.int64)
            image_id = torch.as_tensor([annotations[0].get('image_id')], dtype=torch.int64)
            areas = torch.Tensor(areas)
            iscrowd = torch.zeros((len(annotations),), dtype=torch.uint8)
            masks = None
            return boxes, labels, areas, iscrowd, masks, image_id

        except ValueError as error:
            print(error)

```

```python
class ExecuteSam(object):

    def __init__(self, sam_checkpoint="/content/weights/sam_vit_h_4b8939.pth", model_type="vit_h", device="cuda"):
        try:
            sam = sam_model_registry[model_type](checkpoint=sam_checkpoint)
            sam.to(device=device)
            self.predictor = SamPredictor(sam)
            print('Sam is ok!')
        except ValueError as error:
            print(error)

    def convertCOCO2Sam(self, folder_save='.', json=None, prefix=None, path_mask=None, path_robo=None):
        try:
            extractCoco = ExtractCoco(path=json)
            elements = extractCoco.extract_dic(prefix=prefix)

            path_mask = os.path.join(folder_save, 'masks_max_score') if path_mask is None else path_mask
            path_robo = os.path.join(folder_save, 'other_masks') if path_robo is None else path_robo

            os.makedirs(path_mask, exist_ok=True)
            os.makedirs(path_robo, exist_ok=True)
            os.makedirs(os.path.join(folder_save, 'mascara-combinada'), exist_ok=True)

            for i, element in enumerate(elements):
                print('Loading the image ..:', element[0])
                img = cv2.cvtColor(cv2.imread(element[0]), cv2.COLOR_BGR2RGB)
                boxes = element[1]['boxes']
                labels = [int(e.cpu().numpy()) for e in element[1]['labels']]
                boxes = [[int(b) for b in box.cpu().numpy()] for box in boxes]

                num_classes = np.unique(labels)
                img_processing = img.copy()

                # Iterate over each category in the image
                for category in num_classes:
                    category_boxes = [box for box, label in zip(boxes, labels) if label == category]

                    # Process the image for the current category
                    self.predictor.set_image(img_processing)
                    masks, scores, logits = self.predictor.predict(
                        point_coords=np.array(category_boxes),
                        point_labels=np.array([category] * len(category_boxes)),
                        multimask_output=True,
                    )

                    # Save the masks for the current category in a directory with the category name
                    category_name = extractCoco.dic_labels.get(category, f'category_{category}')
                    category_path_mask = os.path.join(path_mask, category_name)
                    os.makedirs(category_path_mask, exist_ok=True)
                    category_path_robo = os.path.join(path_robo, category_name)
                    os.makedirs(category_path_robo, exist_ok=True)

                    self.save(masks, scores, element[0].split(os.sep)[-1], img, img_processing, folder_save,
                              category_path_mask, category_path_robo)

        except ValueError as error:
            print(error)

    def save(self, masks, scores, file, imageMix, img, folder_save, path_mask, path_robo):
        try:
            # Encontra o √≠ndice da m√°scara com a maior pontua√ß√£o
            max_score_idx = np.argmax(scores)

            # Salva a m√°scara com maior pontua√ß√£o
            max_score_mask = masks[max_score_idx].astype(np.uint8) * 255
            max_score_mask_filename = f"{file}_mascara_max_score.png"
            max_score_mask_path = os.path.join(path_mask, max_score_mask_filename)
            cv2.imwrite(max_score_mask_path, max_score_mask)

            # Salva as m√°scaras com pontua√ß√µes menores em um diret√≥rio separado
            for idx, mask in enumerate(masks):
                if idx != max_score_idx:
                    mask = mask.astype(np.uint8) * 255
                    mask_filename = f"{file}_mascara_{idx}_score_{scores[idx]}.png"
                    mask_path = os.path.join(path_robo, mask_filename)
                    cv2.imwrite(mask_path, mask)

            # Salva a imagem original
            original_image_path = os.path.join(folder_save, 'imagens-originais', file + '.png')
            cv2.imwrite(original_image_path, img)

            # Garante que a m√°scara seja do tipo np.uint8 antes de converter para BGR
            mask = mask.astype(np.uint8)

            # Salva a imagem combinada com a m√°scara de maior pontua√ß√£o
            mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)
            imageMix = cv2.addWeighted(imageMix, 0.5, mask, 0.5, 0)
            combined_image_path = os.path.join(folder_save, 'mascara-combinada', file + '_sobreposicao_max_score_' + str(scores[max_score_idx]) + '.png')
            cv2.imwrite(combined_image_path, imageMix)

        except ValueError as error:
            print(error)
```

```python
class Files(object):

    def __init__(self, local):
        self.local = local

    def lists(self, path, info=[]):
        try:
            for name in os.listdir(path):
                file = os.path.join(path, name)
                if os.path.isdir(file):
                    info = self.lists(file, info)
                elif os.path.isfile(file):
                    if name.replace('.json', '') != name:
                        info.append([file, path])
            return info
        except ValueError as error:
            print(error)

    def extractJson(self):
        try:
            return self.lists(self.local, [])
        except ValueError as error:
            print(error)

```

### **7. Extra√ß√£o das coordenadas das m√°scaras bin√°rias segmentadas.json**



```python
import glob
import json
import os
import cv2

# Label IDs of the dataset representing different categories
category_ids = {
    "Anticarsia_gemmatalis": 1,
    "Coccinellidae": 2,
    "Diabrotica_speciosa": 3,
    "Edessa_meditabunda": 4,
    "Euschistus_heros_adulto": 5,
    "Euschistus_heros_ninfa": 6,
    "Gastropoda": 7,
    "Lagria_villosa": 8,
    "Nezara_viridula_adulto": 9,
    "Nezara_viridula_ninfa": 10,
    "Rhammatocerus_schistocercoides": 11,
    "Spodoptera_albula": 12,
}

MASK_EXT = 'jpg'
ORIGINAL_EXT = 'jpg'
image_id = 0
annotation_id = 0

def images_annotations_info(maskpath):
    global image_id, annotation_id
    annotations = []
    images = []

    # Iterate through categories and corresponding masks
    for category in category_ids.keys():
        for mask_image in glob.glob(os.path.join(maskpath, category, f'*.{MASK_EXT}')):
            original_file_name = os.path.basename(mask_image)
            mask_image_open = cv2.imread(mask_image)

            # Get image dimensions
            height, width, _ = mask_image_open.shape

            # Create or find existing image annotation
            if original_file_name not in map(lambda img: img['file_name'], images):
                image = {
                    "id": image_id + 1,
                    "width": width,
                    "height": height,
                    "file_name": original_file_name,  # Usar o nome original
                }
                images.append(image)
                image_id += 1
            else:
                image = [element for element in images if element['file_name'] == original_file_name][0]

            # Find contours in the mask image
            gray = cv2.cvtColor(mask_image_open, cv2.COLOR_BGR2GRAY)
            _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
            contours = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[0]

            # Create annotation for each contour
            for contour in contours:
                bbox = cv2.boundingRect(contour)
                area = cv2.contourArea(contour)
                segmentation = contour.flatten().tolist()

                annotation = {
                    "iscrowd": 0,
                    "id": annotation_id,
                    "image_id": image['id'],
                    "category_id": category_ids[category],
                    "bbox": bbox,
                    "area": area,
                    "segmentation": [segmentation],
                }

                # Add annotation if area is greater than zero
                if area > 0:
                    annotations.append(annotation)
                    annotation_id += 1

    return images, annotations, annotation_id


def process_masks(mask_path, dest_json):
    global image_id, annotation_id
    image_id = 0
    annotation_id = 0

    # Initialize the COCO JSON format with categories
    coco_format = {
        "info": {},
        "licenses": [],
        "images": [],
        "categories": [{"id": value, "name": key, "supercategory": key} for key, value in category_ids.items()],
        "annotations": [],
    }

    # Create images and annotations sections
    coco_format["images"], coco_format["annotations"], annotation_cnt = images_annotations_info(mask_path)

    # Create the directory if it doesn't exist
    os.makedirs(os.path.dirname(dest_json), exist_ok=True)

    # Save the COCO JSON to a file
    with open(dest_json, "w") as outfile:
        json.dump(coco_format, outfile, sort_keys=True, indent=4)

    print("Created %d annotations for images in folder: %s" % (annotation_cnt, mask_path))

if __name__ == "__main__":
    train_mask_path = "/content/dataset/masks_max_score/"
    train_json_path = "/content/dataset/novo-json/_annotations.coco.json"
    process_masks(train_mask_path, train_json_path)
```

### **8. Convertendo .json para formato compat√≠vel com yolov8**

```python
import json
import os
import shutil
import yaml

# Fun√ß√£o para converter imagens para o formato YOLO
def convert_to_yolo(input_images_path, input_json_path, output_images_path, output_labels_path):
    # Abrir o arquivo JSON contendo as anota√ß√µes das imagens
    with open(input_json_path) as f:
        data = json.load(f)

    # Criar diret√≥rios para imagens e r√≥tulos de sa√≠da
    os.makedirs(output_images_path, exist_ok=True)
    os.makedirs(output_labels_path, exist_ok=True)

    # Lista para armazenar os nomes dos arquivos
    file_names = []
    for filename in os.listdir(input_images_path):
        if filename.endswith(".jpg"):
            source = os.path.join(input_images_path, filename)
            destination = os.path.join(output_images_path, filename)
            shutil.copy(source, destination)
            file_names.append(filename)

    # Fun√ß√£o para obter anota√ß√µes da imagem
    def get_img_ann(image_id):
        return [ann for ann in data['annotations'] if ann['image_id'] == image_id]

    # Fun√ß√£o para obter dados da imagem
    def get_img(filename):
        return next((img for img in data['images'] if img['file_name'] == filename), None)

    # Iterar pelos nomes de arquivo e processar cada imagem
    for filename in file_names:
        img = get_img(filename)
        img_id = img['id']
        img_w = img['width']
        img_h = img['height']
        img_ann = get_img_ann(img_id)

        # Escrever dados normalizados do pol√≠gono em um arquivo de texto
        if img_ann:
            with open(os.path.join(output_labels_path, f"{os.path.splitext(filename)[0]}.txt"), "a") as file_object:
                for ann in img_ann:
                    current_category = ann['category_id'] - 1
                    polygon = ann['segmentation'][0]
                    normalized_polygon = [format(coord / img_w if i % 2 == 0 else coord / img_h, '.6f') for i, coord in enumerate(polygon)]
                    file_object.write(f"{current_category} " + " ".join(normalized_polygon) + "\n")

# Fun√ß√£o para criar um arquivo YAML para o conjunto de dados
def create_yaml(input_json_path, output_yaml_path, train_path, val_path, test_path=None):
    with open(input_json_path) as f:
        data = json.load(f)

    # Extrair os nomes das categorias
    names = [category['name'] for category in data['categories']]

    # N√∫mero de classes
    nc = len(names)

    # Criar um dicion√°rio com o conte√∫do necess√°rio
    yaml_data = {
        'names': names,
        'nc': nc,
        'test': test_path if test_path else '',
        'train': train_path,
        'val': val_path
    }

    # Garantir que o diret√≥rio exista antes de escrever o arquivo YAML
    os.makedirs(os.path.dirname(output_yaml_path), exist_ok=True)

    # Escrever o dicion√°rio em um arquivo YAML
    with open(output_yaml_path, 'w') as file:
        yaml.dump(yaml_data, file, default_flow_style=False)

if __name__ == "__main__":
    base_input_path = "/content/inseto-2/train/"
    base_output_path = "/content/yolo/yolo_dataset/"

    # Processar conjunto de dados de valida√ß√£o (se necess√°rio)
    convert_to_yolo(
        input_images_path=os.path.join(base_input_path, "/content/inseto-2/train/"),
        input_json_path=os.path.join(base_input_path, "/content/dataset/novo-json/_annotations.coco.json"),
        output_images_path=os.path.join(base_output_path, "/content/yolo"),
        output_labels_path=os.path.join(base_output_path, "/content/yolo")
    )

    # # Processar conjunto de dados de treinamento
    # convert_to_yolo(
    #     input_images_path=os.path.join(base_input_path, "train_images"),
    #     input_json_path=os.path.join(base_input_path, "train_images/train.json"),
    #     output_images_path=os.path.join(base_output_path, "train/images"),
    #     output_labels_path=os.path.join(base_output_path, "train/labels")
    # )

    # Criar o arquivo de configura√ß√£o YAML
    create_yaml(
        input_json_path=os.path.join(base_input_path, "/content/dataset/novo-json/_annotations.coco.json"),
        output_yaml_path=os.path.join(base_output_path, "data.yaml"),
        train_path="/content/yolo/yolo_dataset/",
        val_path='../test/images',  # ou None se n√£o aplic√°vel
        test_path='../test/images'  # ou None se n√£o aplic√°vel
    )
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import cv2
import os
import random

def display_random_image_with_annotations(directory, colors=None):
    # Listar todos os arquivos no diret√≥rio
    all_files = os.listdir(directory)

    # Filtrar os arquivos .jpg
    jpg_files = [file for file in all_files if file.endswith('.jpg')]

    if not jpg_files:
        print("Nenhum arquivo .jpg encontrado no diret√≥rio.")
        return

    # Escolher aleatoriamente um arquivo .jpg
    random_jpg = random.choice(jpg_files)

    # Verificar se existe um arquivo .txt correspondente
    txt_file = os.path.splitext(random_jpg)[0] + '.txt'

    if txt_file not in all_files:
        print(f"Arquivo .txt correspondente n√£o encontrado para {random_jpg}.")
        return

    # Chamar a fun√ß√£o existente para exibir a imagem com anota√ß√µes
    image_path = os.path.join(directory, random_jpg)
    annotation_path = os.path.join(directory, txt_file)
    display_image_with_annotations(image_path, annotation_path, colors)

def display_image_with_annotations(image_path, annotation_path, colors=None):
    # Load image using OpenCV and convert it from BGR to RGB color space
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    img_h, img_w, _ = image.shape

    # Create a figure and axis to display the image
    fig, ax = plt.subplots(1)
    ax.imshow(image)
    ax.axis('off')  # Turn off the axes

    # Define a default color map if none is provided
    if colors is None:
        colors = plt.cm.get_cmap('tab10')

    # Open the annotation file and process each line
    with open(annotation_path, 'r') as file:
        for line in file:
            parts = line.strip().split()
            category_id = int(parts[0])
            # Choose color based on category ID, looping through color map if more than 10 categories
            color = colors(category_id % 10)
            # Extract normalized polygon coordinates and denormalize them
            polygon = [float(coord) for coord in parts[1:]]
            polygon = [coord * img_w if i % 2 == 0 else coord * img_h for i, coord in enumerate(polygon)]
            # Reshape into (num_points, 2) array
            polygon = [(polygon[i], polygon[i+1]) for i in range(0, len(polygon), 2)]
            # Create a Polygon patch using the denormalized coordinates
            patch = patches.Polygon(polygon, closed=True, edgecolor=color, fill=False)
            # Add the patch to the plot to display the annotated region
            ax.add_patch(patch)

    plt.show()  # Display the image with annotations

directory_path = "/content/yolo/"
display_random_image_with_annotations(directory_path)
```

### **3. k-dobras**

```python
import os
import random
from shutil import copyfile

def create_folds_with_annotations(origem_dataset, destino_dobras, num_folds):
    # Lista de arquivos no diret√≥rio original
    arquivos_jpg = [arquivo for arquivo in os.listdir(origem_dataset) if arquivo.endswith('.jpg')]
    random.shuffle(arquivos_jpg)  # Embaralha a lista para sele√ß√£o aleat√≥ria

    # Calcula o tamanho do subconjunto para cada dobra
    tamanho_subconjunto = len(arquivos_jpg) // num_folds

    # Divide a lista de arquivos em k dobras
    dobras = [arquivos_jpg[i * tamanho_subconjunto: (i + 1) * tamanho_subconjunto] for i in range(num_folds)]

    # Loop sobre cada dobra
    for i, dobra in enumerate(dobras):
        # Cria um diret√≥rio para a dobra
        pasta_dobra = os.path.join(destino_dobras, f'dobra_{i + 1}')
        os.makedirs(pasta_dobra, exist_ok=True)

        # Copia os arquivos .jpg e .txt correspondentes para a dobra
        for arquivo_jpg in dobra:
            # Constr√≥i o nome do arquivo .txt correspondente
            arquivo_txt = arquivo_jpg.replace('.jpg', '.txt')

            # Verifica se o arquivo .txt existe antes de copiar
            if os.path.exists(os.path.join(origem_dataset, arquivo_txt)):
                origem_arquivo_jpg = os.path.join(origem_dataset, arquivo_jpg)
                destino_arquivo_jpg = os.path.join(pasta_dobra, arquivo_jpg)
                copyfile(origem_arquivo_jpg, destino_arquivo_jpg)

                origem_arquivo_txt = os.path.join(origem_dataset, arquivo_txt)
                destino_arquivo_txt = os.path.join(pasta_dobra, arquivo_txt)
                copyfile(origem_arquivo_txt, destino_arquivo_txt)

        print(f"Arquivos movidos para a dobra {i + 1} em {pasta_dobra}")

if __name__ == "__main__":
    # Define os diret√≥rios de origem e destino
    origem_dataset = "/content/yolo/yolo_dataset"
    destino_dobras = "/content/yolo/dobras"
    num_folds = 5

    # Cria as dobras com os arquivos correspondentes
    create_folds_with_annotations(origem_dataset, destino_dobras, num_folds)
```

Criando .yaml para cada dobra.

```python
import os
import yaml

# Caminho para as pastas das dobras
dobras_path = '/content/yolo/dobras/'

# Lista de extens√µes de arquivos suportadas
supported_extensions = ['.jpg', '.jpeg', '.png', '.txt']

# Loop pelas dobras
for dobra in os.listdir(dobras_path):
    dobra_path = os.path.join(dobras_path, dobra)

    # Verifica se √© um diret√≥rio
    if os.path.isdir(dobra_path):
        # Lista de arquivos de imagem e r√≥tulos
        files = [file for file in os.listdir(dobra_path) if any(file.endswith(ext) for ext in supported_extensions)]

        # Separa as imagens e r√≥tulos
        images = [file for file in files if file.endswith('.jpg') or file.endswith('.jpeg') or file.endswith('.png')]
        labels = [file for file in files if file.endswith('.txt')]

        # Caminho para o arquivo YAML
        yaml_path = os.path.join(dobra_path, f'{dobra}.yaml')

        # Cria√ß√£o do arquivo YAML
        with open(yaml_path, 'w') as yaml_file:
            data = {
                'path': dobra_path,
                'train': 'train',
                'val': 'val',
                'names': ["Anticarsia_gemmatalis", "Coccinellidae", "Diabrotica_speciosa", "Edessa_meditabunda", "Euschistus_heros_adulto", "Euschistus_heros_ninfa", "Gastropoda", "Lagria_villosa", "Nezara_viridula_adulto", "Nezara_viridula_ninfa", "Rhammatocerus_schistocercoides", "Spodoptera_albula"],  # Substitua com suas classes reais
                'train_images': [os.path.join(dobra_path, img) for img in images],
                'train_labels': [os.path.join(dobra_path, label) for label in labels]
            }
            yaml.dump(data, yaml_file)
```

Divindo arquivos entre train e val

```python
import os
import random
import shutil

dobras_path = '/content/yolo/dobras/'
train_percentage = 0.7

for dobra in os.listdir(dobras_path):
    dobra_path = os.path.join(dobras_path, dobra)

    if os.path.isdir(dobra_path):
        print(f"Dividindo dados em {dobra}...")

        train_dobra_dir = os.path.join(dobra_path, 'train')
        val_dobra_dir = os.path.join(dobra_path, 'val')
        os.makedirs(train_dobra_dir, exist_ok=True)
        os.makedirs(val_dobra_dir, exist_ok=True)

        files = os.listdir(dobra_path)

        for filename in files:
            if filename.endswith(".jpg"):
                if random.random() < train_percentage:
                    dest_dir = train_dobra_dir
                else:
                    dest_dir = val_dobra_dir

                img_src_path = os.path.join(dobra_path, filename)
                txt_src_path = os.path.join(dobra_path, filename.replace(".jpg", ".txt"))
                img_dest_path = os.path.join(dest_dir, filename)
                txt_dest_path = os.path.join(dest_dir, filename.replace(".jpg", ".txt"))

                shutil.move(img_src_path, img_dest_path)
                shutil.move(txt_src_path, txt_dest_path)

        print(f"Divis√£o conclu√≠da para {dobra}.\n")

print("Divis√£o conclu√≠da para todos os subdiret√≥rios.")
```
